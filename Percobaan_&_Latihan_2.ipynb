```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
```

```python
class SinglyLinkedList:
  def __init__(self):
    self.head = None

  def insert_at_end(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      return
    temp = self.head
    while temp.next:
      temp = temp.next
    temp.next = new_node

  def display(self):
    temp = self.head
    while temp:
      print(temp.data, end=' -> ')
      temp = temp.next
    print('None')
```

```python
class DoublyNode:
  def __init__(self, data):
    self.data = data
    self.next = None
    self.prev = None

class DoublyLinkedList:
  def __init__(self):
    self.head = None

  def insert_at_end(self, data):
    new_node = DoublyNode(data)
    if not self.head:
      self.head = new_node
      return
    temp = self.head
    while temp.next:
      temp = temp.next
    temp.next = new_node
    new_node.prev = temp

  def display(self):
    temp = self.head
    while temp:
      print(temp.data, end=' <-> ')
      temp = temp.next
    print('None')
```

```python
class CircularLinkedList:
  def __init__(self):
    self.head = None

  def insert_at_end(self, data):
    new_node = Node(data)
    if not self.head:
      self.head = new_node
      new_node.next = self.head

  def display(self):
    if not self.head:
      return
    temp = self.head
    while True:
      print(temp.data, end=' -> ')
      temp = temp.next
      if temp == self.head:
        break
    print('(Back to Head)')
```

```python
# Implementasi Singly Linked List
class TaskNode:
  def __init__(self, task_name):
    self.task_name = task_name  # Menyimpan nama tugas
    self.next = None # Pointer ke tugas berikutnya

# Definisi Linked List untuk Daftar Tugas
class TaskList:
  def __init__(self):
    self.head = None # Awalnya tidak ada tugas (daftar kosong)

  # Menambah tugas ke Linked List
  def add_task(self, task_name):
    new_task = TaskNode(task_name)
    if not self.head:
      self.head = new_task # Jika list kosong, tugas pertama menjadi head
    else:
        current = self.head
        while current.next: # Mencari node terakhir dalam list
          current = current.next
        current.next = new_task # Menambah tugas di akhir list

  # Menampilkan daftar tugas
  def show_tasks(self):
    if not self.head:
      print('\n Tidak ada tugas hari ini')
      return

    print('\n Daftar Tugas Hari Ini:')
    current = self.head
    count = 1
    while current:
      print(f"{count}. {current.task_name}")
      current = current.next
      count += 1

# ====== Program Utama(Main) ======

# Membuat objek daftar tugas
task_list = TaskList()

# Meminta pengguna memasukkan jumlah tugas
num_tasks = int(input('Masukkan jumlah tugas hari ini: '))

# Loop untuk memasukkan tugas ke dalam Linked List
for i in range(1, num_tasks + 1):
  task_name = input(f'Masukkan tugas ke-{i}: ')
  task_list.add_task(task_name)

# Menampilkan hasil daftar tugas
task_list.show_tasks()
```

```python
# Latihan 1
class TaskNode:
    def __init__(self, task_name):
        self.task_name = task_name
        self.next = None

# Definisi Linked List untuk Daftar Mata Kuliah
class TaskList:
    def __init__(self):
        self.head = None

    # Menambah mata kuliah ke Linked List
    def add_task(self, task_name):
        new_task = TaskNode(task_name)
        if not self.head:
            self.head = new_task
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_task

    # Menampilkan daftar mata kuliah
    def show_tasks(self):
        if not self.head:
            print('\nTidak ada mata kuliah pada hari Jumat')
            return

        print('\nJadwal Mata Kuliah Hari Jumat:')
        current = self.head
        count = 1
        while current:
            print(f"{count}. {current.task_name}")
            current = current.next
            count += 1

# ====== Program Utama (Main) ======

# Membuat objek daftar mata kuliah
task_list = TaskList()

# Meminta pengguna memasukkan nama mata kuliah
num_tasks = 4  # Jumlah mata kuliah pada hari Jumat

print(f"Masukkan {num_tasks} mata kuliah yang dijadwalkan pada hari Jumat:")

# Loop untuk memasukkan nama mata kuliah ke dalam Linked List
for i in range(1, num_tasks + 1):
    task_name = input(f'Masukkan mata kuliah ke-{i}: ')
    task_list.add_task(task_name)

# Menampilkan hasil daftar mata kuliah pada hari Jumat
task_list.show_tasks()
```

```python
# Definisi Node dalam Doubly Linked List
class PageNode:
  def __init__(self, url):
    self.url = url
    self.next = None
    self.prev = None

# Definisi Doubly Linked List untuk Riwayat Peramban
class BrowseHistory:
  def __init__(self):
    self.current = None

  # Menambahkan halaman baru ke dalam riwayat
  def visit_page(self, url):
    new_page = PageNode(url)
    if not self.current:
      self.current = new_page
    else:
      new_page.prev = self.current
      self.current.next = new_page
      self.current = new_page

  # Menampilkan riwayat kunjungan
  def show_history(self):
    if not self.current:
      print('\nRiwayat Peramban Web: ')
      temp = self.current
      while temp.prev:
        temp = temp.prev

      count = 1
      while temp:
        print(f'{count}. {temp.url}')
        temp = temp.next
        count += 1

  # Navigasi mundur ke halaman sebelumnya
  def go_back(self):
    if self.current and self.current.prev:
      self.current = self.current.prev
      print(f'\nAnda kembali ke: {self.current.url}')
    else:
      print('\nTidak bisa kembali, ini adalah halaman pertama!')

  # Navigasi maju ke halaman berikutnya
  def go_forward(self):
    if self.current and self.current.next:
      self.current = self.current.next
      print(f'\nAnda maju ke: {self.current.url}')
    else:
      print('\nTidak bisa maju, ini adalah halaman terakhir!')

# ====== Program Utama (Main) ====== #

# Membuat objek riwayat peramban
history = BrowseHistory()

# Meminta pengguna memasukkan jumlah halaman yang dikunjungi
num_pages = int(input('\nMasukkan jumlah halaman yang dikunjungi: '))

# Loop untuk memasukkan url halaman ke dalam Doubly Linked List
for i in range(1, num_pages + 1):
  url = input(f'Masukkan URL halaman ke-{i}: ')
  history.visit_page(url)

# Menampilkan riwayat yang telah dimasukkan
history.show_history()

# Menu navigasi pengguna
while True:
  command = input('\nTekan "b" untuk kembalin Tekan "f" untuk maju, Tekan "x" untuk keluar. \nMasukkan perintah: '). lower()

  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print('\n Keluar dari riwayat peramban')
    break
  else:
    print('\nPerintah tidak dikenal.')
```

```python
# Latihan 2

# Definisi Node dalam Doubly Linked List
class FolderNode:
    def __init__(self, folder_name):
        self.folder_name = folder_name
        self.next = None
        self.prev = None

# Definisi Doubly Linked List untuk Riwayat Navigasi Folder
class FolderHistory:
    def __init__(self):
        self.current = None

    # Menambahkan folder baru ke dalam riwayat
    def visit_folder(self, folder_name):
        new_folder = FolderNode(folder_name)
        if not self.current:
            self.current = new_folder
        else:
            new_folder.prev = self.current
            self.current.next = new_folder
            self.current = new_folder

    # Menampilkan riwayat navigasi folder
    def show_history(self):
        if not self.current:
            print('\nTidak ada riwayat folder yang dikunjungi')
            return

        print('\nRiwayat Navigasi Folder:')
        temp = self.current
        count = 1

        # Menampilkan folder dari folder terakhir yang dikunjungi
        while temp.prev:
            temp = temp.prev

        # Menampilkan semua folder dari awal sampai akhir
        while temp:
            print(f"{count}. {temp.folder_name}")
            temp = temp.next
            count += 1

    # Navigasi mundur ke folder sebelumnya
    def go_back(self):
        if self.current and self.current.prev:
            self.current = self.current.prev
            print(f'\nAnda kembali ke folder: {self.current.folder_name}')
        else:
            print('\nTidak bisa kembali, ini adalah folder pertama!')

    # Navigasi maju ke folder berikutnya
    def go_forward(self):
        if self.current and self.current.next:
            self.current = self.current.next
            print(f'\nAnda maju ke folder: {self.current.folder_name}')
        else:
            print('\nTidak bisa maju, ini adalah folder terakhir!')

# ====== Program Utama ====== #

# Membuat objek riwayat navigasi folder
folder_history = FolderHistory()

# Meminta pengguna memasukkan nama folder yang dikunjungi
num_folders = int(input('\nMasukkan jumlah folder yang akan dikunjungi: '))

# Loop untuk memasukkan folder ke dalam Doubly Linked List
for i in range(1, num_folders + 1):
    folder_name = input(f'Masukkan folder ke-{i}: ')
    folder_history.visit_folder(folder_name)

# Menampilkan riwayat navigasi folder
folder_history.show_history()

# Menu navigasi pengguna
while True:
    command = input('\nTekan "b" untuk kembali, Tekan "f" untuk maju, Tekan "x" untuk keluar. \nMasukkan perintah: ').lower()

    if command == 'b':
        folder_history.go_back()
    elif command == 'f':
        folder_history.go_forward()
    elif command == 'x':
        print('\nKeluar dari riwayat navigasi folder')
        break
    else:
        print('\nPerintah tidak dikenal.')
```

```python
# Definisi Node dalam Circular Linked List
class SongNode:
  def __init__(self, title):
    self.title = title
    self.next = None

# Definisi Circular Linked List untuk Playlist Lagu
class Playlist:
  def __init__(self):
    self.current = None

  # Menambahkan lagu ke dalam playlist
  def add_song(self, title):
    new_song = SongNode(title)
    if not self.current:
      self.current = new_song
      self.current.next = self.current
    else:
      temp = self.current
      while temp.next != self.current:
        temp = temp.next
      temp.next = new_song
      new_song.next = self.current

  # Menampilkan daftar lagu dalam playlist
  def show_playlist(self):
    if not self.current:
      print('\nPlaylist kosong!')
      return

    print('\nPlaylist Lagu Anda: ')
    temp = self.current
    count = 1
    while True:
      print(f'{count}. {temp.title}')
      temp = temp.next
      count += 1
      if temp == self.current:
        break

  # Memutar lagu berikutnya dalam playlist
  def next_song(self):
    if self.current:
      self.current = self.current.next
      print(f'\n Memutar: {self.current.title}')

# ====== Program Utama (Main) ====== #

# Membuat objek playlist
playlist = Playlist()

# Meminta pengguna memasukkan jumlah lagu
num_songs = int(input('\nMasukkan jumlah lagu dalam playlist: '))

# Loop untuk memasukkan lagu ke dalam Circular Linked List
for i in range(1, num_songs + 1):
  title = input(f'Masukkan judul lagu ke-{i}: ')
  playlist.add_song(title)

# Menampilkan daftar lagu
playlist.show_playlist()

#Menu navigasi pengguna
while True:
  command = input('\nTekan "n" untuk lagu berikutnya, Tekan "x" untuk keluar. \nMasukkan perintah: ').lower()
  if command == 'n':
    playlist.next_song()
  elif command == 'x':
    print('\nKeluar dari pemutar musik.')
    break
  else:
    print('\nPerintah tidak dikenal.')
```

```python
# Operasi Dasar Pada Linked List
class Node:
    def __init__(self, data):
        self.data = data  # Menyimpan nilai node
        self.next = None  # Pointer ke node berikutnya


class LinkedList:
    def __init__(self):
        self.head = None  # Awalnya, daftar kosong

    # Menambahkan elemen di awal
    def add_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    # Menambahkan elemen di akhir
    def add_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Menambahkan elemen di posisi tertentu
    def add_at_position(self, position, data):
        if position < 0:
            print("Posisi tidak valid!")
            return
        new_node = Node(data)
        if position == 0:
            self.add_at_beginning(data)
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp:
                print("Posisi melebihi panjang daftar!")
                return
            temp = temp.next
        new_node.next = temp.next
        temp.next = new_node

    # Menghapus elemen di awal
    def delete_at_beginning(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        self.head = self.head.next

    # Menghapus elemen di akhir
    def delete_at_end(self):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if not self.head.next:
            self.head = None
            return
        temp = self.head
        while temp.next.next:
            temp = temp.next
        temp.next = None

    # Menghapus elemen di posisi tertentu
    def delete_at_position(self, position):
        if not self.head:
            print("Daftar kosong, tidak ada yang bisa dihapus!")
            return
        if position == 0:
            self.head = self.head.next
            return
        temp = self.head
        for _ in range(position - 1):
            if not temp or not temp.next:
                print("Posisi tidak ditemukan!")
                return
            temp = temp.next
        temp.next = temp.next.next

    # Mencari elemen dalam Linked List
    def search(self, key):
        temp = self.head
        while temp:
            if temp.data == key:
                print(f"Elemen {key} ditemukan dalam daftar!")
                return
            temp = temp.next
        print(f"Elemen {key} tidak ditemukan dalam daftar.")

    # Membalik (reverse) Linked List
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    # Menampilkan Linked List
    def display(self):
        if not self.head:
            print("Daftar kosong!")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")


# Program Utama (Main)
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("\nMasukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
    data = int(input(f"Masukkan data ke-{i+1}: "))
    ll.add_at_end(data)

# Menu Operasi Linked List
while True:
    print("\nPilih Operasi:")
    print("1. Tambah elemen di awal")
    print("2. Tambah elemen di akhir")
    print("3. Tambah elemen di posisi tertentu")
    print("4. Hapus elemen di awal")
    print("5. Hapus elemen di akhir")
    print("6. Hapus elemen di posisi tertentu")
    print("7. Cari elemen dalam daftar")
    print("8. Balikkan daftar")
    print("9. Tampilkan daftar")
    print("0. Keluar")

    choice = int(input("\nMasukkan pilihan: "))

    if choice == 1:
        data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
        ll.add_at_beginning(data)
    elif choice == 2:
        data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
        ll.add_at_end(data)
    elif choice == 3:
        position = int(input("Masukkan posisi untuk menambahkan data: "))
        data = int(input("Masukkan data yang ingin ditambahkan: "))
        ll.add_at_position(position, data)
    elif choice == 4:
        ll.delete_at_beginning()
    elif choice == 5:
        ll.delete_at_end()
    elif choice == 6:
        position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
        ll.delete_at_position(position)
    elif choice == 7:
        key = int(input("Masukkan elemen yang ingin dicari: "))
        ll.search(key)
    elif choice == 8:
        ll.reverse()
        print("Daftar berhasil dibalik!")
    elif choice == 9:
        ll.display()
    elif choice == 0:
        print("\nKeluar dari program.")
        break
    else:
        print("\nPilihan tidak valid, silakan coba lagi!")
```
