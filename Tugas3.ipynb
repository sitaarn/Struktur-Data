```python
class MenaraHanoi:
    def __init__(self, jumlah_cakram, langkah_max):
        self.jumlah_cakram = jumlah_cakram
        self.langkah_max = langkah_max
        self.langkah = 0
        self.tiang = {
            'A': list(reversed(range(1, jumlah_cakram + 1))),
            'B': [],
            'C': []
        }

    def pindah_cakram(self, asal, tujuan):
        if self.langkah >= self.langkah_max:
            print("Batas langkah tercapai! Permainan berakhir.")
            return

        if not self.tiang[asal]:
            print(f"Tiang {asal} kosong! Tidak ada cakram untuk dipindahkan.")
            return

        if self.tiang[tujuan] and self.tiang[tujuan][-1] < self.tiang[asal][-1]:
            print("Tidak bisa meletakkan cakram besar di atas cakram kecil!")
            return

        cakram = self.tiang[asal].pop()
        self.tiang[tujuan].append(cakram)
        self.langkah += 1
        print(f"Memindahkan cakram {cakram} dari {asal} ke {tujuan}")

    def tampilkan_tiang(self):
        print("\nStatus Tiang:")
        for nama, isi in self.tiang.items():
            print(f"{nama}: {isi}")

    def selesai(self):
        return len(self.tiang['C']) == self.jumlah_cakram

# Mulai Permainan
print("Selamat datang di permainan Menara Hanoi!")
jumlah_cakram = int(input('Masukkan Jumlah Cakram: '))
langkah_max = int(input('Masukkan Langkah Max: '))

game = MenaraHanoi(jumlah_cakram, langkah_max)
game.tampilkan_tiang()

while not game.selesai() and game.langkah < game.langkah_max:
    print("\nPilih langkah Anda:")
    asal = input("Pindahkan dari tiang (A/B/C): ").upper()
    tujuan = input("Ke tiang (A/B/C): ").upper()

    if asal in 'ABC' and tujuan in 'ABC' and asal != tujuan:
        game.pindah_cakram(asal, tujuan)
        game.tampilkan_tiang()
    else:
        print("Input tidak valid! Masukkan hanya A, B, atau C dengan tujuan yang berbeda.")

if game.selesai():
    print("Selamat! Anda berhasil menyelesaikan Menara Hanoi!")
else:
    print("Langkah habis! Coba lagi nanti.")
```
